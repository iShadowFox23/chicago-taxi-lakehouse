## **Paso 2: Crear el cliente de BigQuery**

```python
# Importa la librería oficial de BigQuery, permite que Python se comunique
# con BigQuery mediante la API de Google Cloud
from google.cloud import bigquery

# Ingresa el ID de nuestro proyecto en Google Cloud
project_id = 'gestiondatos-chicago'

# Se crea un objeto cliente de BigQuery que permite ejecutar
# ejecutar consultas SQL
client = bigquery.Client(project=project_id)
```

## **Paso 3: Definir las consultas SQL**

```python
# Consulta SQL de todos los datos del dataset de nuestro proyecto
year = 2022

query = f"""
SELECT *
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
WHERE EXTRACT(YEAR FROM trip_start_timestamp) = {year}
"""
```

## **Paso 4: Ejecutar la consulta**

```python
# Enviamos la consulta a Big Query para que se ejecute
query_job = client.query(query)
```

## **Paso 5: Obtener los resultados**

```python
# Se devuelve el resultado de la consulta en BigQuery
results = query_job.result()
```

## **Paso 6: Visualizar los resultados**

```python
# Iteramos fila por fila sobre los resultados de la consulta
contador = 0

for row in results:
  print(row)

  contador += 1

  if contador == 10:
    break
```

## **Paso 7: Convertir el resultdo a DataFrame de Pandas**

```python
df = query_job.to_dataframe()
df.head()
```

## **Paso 8: Exportar a formato Parquet para Descargarlo**

```python
# Hacemos otra consulta con todos los datos del año 2017
df.to_parquet('chicago_taxi_trips2021.parquet')
```
